```python
def load_dataset(path, rename_map, date_columns=None, file_format="json"):
    if file_format == "csv":
        df = pd.read_csv(path)
    else:
        df = pd.read_json(path)

    df.rename(columns=rename_map, inplace=True)

    for column in date_columns or []:
        df[column] = pd.to_datetime(df[column])

    return df
```

# Database design 

```python
import os
if os.path.exists("project.db"):
    os.remove("project.db")
```

```python
conn = sqlite3.connect("project.db")
cursor = conn.cursor()
conn.execute("PRAGMA foreign_keys = ON;")
```

```python
def run_query(query):
    return pd.read_sql(query, conn)


def create_table(table_name, create_query, df):
    cursor.execute(create_query)
    df.to_sql(table_name, conn, if_exists="append", index=False)
    conn.commit()


def create_index(index_name, table, column):
    cursor.execute(f"CREATE INDEX IF NOT EXISTS {index_name} ON {table}({column});")
```

```python
# Query to create the agents table
agents_table_query = """
CREATE TABLE IF NOT EXISTS agents(
    agentid TEXT PRIMARY KEY,
    name TEXT NOT NULL,
    phone TEXT NOT NULL,
    email TEXT NOT NULL,
    commissionrate REAL NOT NULL,
    dealsclosed INTEGER NOT NULL,
    rating REAL NOT NULL,
    experienceyears INTEGER NOT NULL,
    avgclosingdays INTEGER NOT NULL
);
"""
create_table("agents", agents_table_query, agents_df)
```

```python
agents_select_query = "SELECT * FROM agents LIMIT 1;"
run_query(agents_select_query)
```

```python
# Query to create the listings table
listings_table_query = """
CREATE TABLE IF NOT EXISTS listings(
    listingid TEXT PRIMARY KEY,
    city TEXT NOT NULL,
    propertytype TEXT NOT NULL,
    price REAL NOT NULL,
    sqft REAL NOT NULL,
    datelisted DATE NOT NULL,
    agentid TEXT NOT NULL,
    latitude REAL NOT NULL,
    longitude REAL NOT NULL,

    FOREIGN KEY(agentid)
        REFERENCES agents(agentid)
);
"""
create_table("listings", listings_table_query, listings_df)
```

```python
listings_select_query = "SELECT * FROM listings LIMIT 1;"
run_query(listings_select_query)
```

```python
# Query to create the sales table
sales_table_query = """
CREATE TABLE IF NOT EXISTS sales(
    listingid TEXT PRIMARY KEY,
    saleprice REAL NOT NULL,
    datesold DATE NOT NULL,
    daysonmarket INTEGER NOT NULL,

    FOREIGN KEY(listingid)
        REFERENCES listings(listingid)
);
"""
create_table("sales", sales_table_query, sales_df)
```

```python
sales_select_query = "SELECT * FROM sales LIMIT 1;"
run_query(sales_select_query)
```

```python
# Query to create the buyers table
buyers_table_query = """
CREATE TABLE IF NOT EXISTS buyers(
    buyerid INTEGER PRIMARY KEY,
    saleid TEXT NOT NULL,

    buyertype TEXT NOT NULL,
    paymentmode TEXT NOT NULL,
    loantaken BOOLEAN NOT NULL,
    loanprovider TEXT,
    loanamount REAL NOT NULL,

    FOREIGN KEY(saleid)
        REFERENCES sales(listingid)
);
"""
create_table("buyers", buyers_table_query, buyers_df)
```

```python
buyers_select_query = "SELECT * FROM buyers LIMIT 1;"
run_query(buyers_select_query)
```

```python
# Query to create the property_attributes table
property_attributes_table_query = """
CREATE TABLE IF NOT EXISTS property_attributes(
    attributeid INTEGER PRIMARY KEY,
    listingid TEXT UNIQUE,
    bedrooms INTEGER NOT NULL,
    bathrooms INTEGER NOT NULL,
    floornumber INTEGER NOT NULL,
    totalfloors INTEGER NOT NULL,
    yearbuilt INTEGER NOT NULL,
    isrented BOOLEAN NOT NULL,
    tenantcount INTEGER NOT NULL,
    furnishingstatus TEXT NOT NULL,
    metrodistancekm REAL NOT NULL,
    parkingavailable BOOLEAN NOT NULL,
    powerbackup BOOLEAN NOT NULL,

    FOREIGN KEY(listingid)
        REFERENCES listings(listingid)
);
"""
create_table("property_attributes", property_attributes_table_query, property_df)
```

```python
property_select_query = "SELECT * FROM property_attributes LIMIT 1;"
run_query(property_select_query)
```

```python
indexes = [
    ("idx_agents_rating", "agents", "rating"),
    ("idx_agents_experience", "agents", "experienceyears"),
    ("idx_listings_city", "listings", "city"),
    ("idx_listings_propertytype", "listings", "propertytype"),
    ("idx_listings_price", "listings", "price"),
    ("idx_listings_agentid", "listings", "agentid"),
    ("idx_sales_datesold", "sales", "datesold"),
    ("idx_sales_saleprice", "sales", "saleprice"),
    ("idx_buyers_saleid", "buyers", "saleid"),
    ("idx_buyers_buyertype", "buyers", "buyertype"),
    ("idx_buyers_paymentmode", "buyers", "paymentmode"),
    ("idx_property_bedrooms", "property_attributes", "bedrooms"),
    ("idx_property_yearbuilt", "property_attributes", "yearbuilt"),
    ("idx_property_metrodistance", "property_attributes", "metrodistancekm"),
    ("idx_property_listingid", "property_attributes", "listingid"),
]

for index_name, table, column in indexes:
    create_index(index_name, table, column)

conn.commit()
print("All indexes created successfully.")
```

## Property & Pricing Analysis

### 1. What is the average listing price by city?

```python
query1 = """
SELECT city, ROUND(AVG(price), 2) AS average_listing_price
FROM listings
GROUP BY city
ORDER BY average_listing_price DESC;
"""
result1 = run_query(query1)
display(result1)
interpretation = """
New York has the highest average listing price at approximately 2.49 million, followed closely by Phoenix,
Los Angeles, and Houston. Chicago has the lowest average listing price. The spread across cities is narrow
(~62K, roughly 2.5% of the average), suggesting listing prices are fairly consistent across all five cities
in this dataset.
"""
print(interpretation)
```

### 2. What is the average price per square foot by property type?

```python
query2 = """
SELECT propertytype, ROUND(AVG(price / sqft), 2) AS avg_price_per_sqft
FROM listings
GROUP BY propertytype
ORDER BY avg_price_per_sqft DESC;
"""
result2 = run_query(query2)
display(result2)
interpretation = """
Houses have the highest average price per square foot at ~796, followed closely by apartments (792) and townhouses (790). 
Condos are the cheapest per square foot at ~755, roughly 5% below the group average. 
Overall, the spread across property types is fairly narrow (~41 per sqft, about 5% of the average), 
suggesting price-per-sqft is not a strong differentiator between property types like house, apartment, and townhouse cluster tightly, with condo pulling the average down slightly."""
print(interpretation)
```

### 3. How does furnishing status impact property prices?

```python
query3 = """
SELECT p.furnishingstatus, ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p ON l.listingid = p.listingid
GROUP BY p.furnishingstatus
ORDER BY average_price DESC;
"""
result3 = run_query(query3)
display(result3)
interpretation = """
Interpretation
* Furnished properties have the highest average price.
* Semi-Furnished properties are almost identical in price (difference of only about 2,346).
* Unfurnished properties have the lowest average price.

Observation
The difference is very small:
Furnished vs Semi-Furnished -> ~2.3K
Furnished vs Unfurnished -> ~29.6K
"""
print(interpretation)
```

### 4. Do properties closer to metro stations command higher prices?

```python
query4 = """
SELECT
    CASE
        WHEN p.metrodistancekm <= 1 THEN 'Within 1 km'
        WHEN p.metrodistancekm <= 3 THEN '1 - 3 km'
        WHEN p.metrodistancekm <= 5 THEN '3 - 5 km'
        ELSE 'More than 5 km'
    END AS metro_distance,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p ON l.listingid = p.listingid
GROUP BY metro_distance
ORDER BY average_price DESC;
"""
result4 = run_query(query4)
display(result4)
interpretation = """
Interpretation:
* Properties more than 5 km from a metro station have the highest average price.
* Properties within 1 km have the lowest average price.
The difference between the 1-3 km and 3-5 km groups is quite small.

Conclusion
No. This dataset does not indicate that properties closer to metro stations command higher prices.
In fact, properties located more than 5 km away have the highest average listing price.
"""
print(interpretation)
```

### 5. Are rented properties priced differently from non-rented ones?

```python
query5 = """
SELECT
    CASE
        WHEN p.isrented = 1 THEN 'Rented'
        ELSE 'Non Rented'
    END AS rental_status,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p ON l.listingid = p.listingid
GROUP BY rental_status
ORDER BY average_price DESC;
"""
result5 = run_query(query5)
display(result5)
interpretation = """
Interpretation:
The average prices of rented and non-rented properties are almost identical.
This suggests that rental status alone does not have a significant impact on property prices in this dataset.
"""
print(interpretation)
```

### 6. How do bedrooms and bathrooms affect pricing?

```python
query6 = """
SELECT p.bedrooms, p.bathrooms, COUNT(*) AS total_properties, ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p ON l.listingid = p.listingid
GROUP BY p.bedrooms, p.bathrooms
ORDER BY p.bedrooms, p.bathrooms;
"""
result6 = run_query(query6)
display(result6)
interpretation = """
The number of bedrooms and bathrooms alone does not appear to be a strong predictor of property price in this dataset.
Property prices likely depend on a combination of factors such as city, property type, square footage, furnishing status,
and location rather than just room counts.
"""
print(interpretation)
```

### 7. Do properties with parking and power backup sell at higher prices?

```python
query7 = """
SELECT
    CASE
        WHEN p.parkingavailable = 1 THEN 'Parking'
        ELSE 'No Parking'
    END AS parking,
    CASE
        WHEN p.powerbackup = 1 THEN 'Power Backup'
        ELSE 'No Power Backup'
    END AS power_backup,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p ON l.listingid = p.listingid
GROUP BY p.parkingavailable, p.powerbackup
ORDER BY average_price DESC;
"""
result7 = run_query(query7)
display(result7)
interpretation = """
Interpretation:
* Properties with both parking and power backup have the highest average price.
* Properties with no parking and no power backup have the lowest average price.
Conclusion:
Properties that offer both parking and power backup tend to have the highest average prices.
Although the price differences are modest, these amenities appear to add value and are associated with
slightly higher-priced properties.
"""
print(interpretation)
```

### 8. How does year built influence listing price?

```python
query8 = """
SELECT
    CASE
        WHEN p.yearbuilt BETWEEN 1990 AND 1995 THEN 'Between 1990 and 1995'
        WHEN p.yearbuilt BETWEEN 1996 AND 2000 THEN 'Between 1996 and 2000'
        WHEN p.yearbuilt BETWEEN 2001 AND 2005 THEN 'Between 2001 and 2005'
        WHEN p.yearbuilt BETWEEN 2006 AND 2010 THEN 'Between 2006 and 2010'
        WHEN p.yearbuilt BETWEEN 2011 AND 2015 THEN 'Between 2011 and 2015'
        WHEN p.yearbuilt BETWEEN 2016 AND 2020 THEN 'Between 2016 and 2020'
        WHEN p.yearbuilt BETWEEN 2021 AND 2025 THEN 'Between 2021 and 2025'
        ELSE 'Other'
    END AS built_between,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p ON l.listingid = p.listingid
GROUP BY built_between
ORDER BY average_price DESC;
"""
result8 = run_query(query8)
display(result8)
interpretation = """
Observations
* The oldest properties (1990-1995) have the highest average listing price.
* The newest properties (2021-2025) rank second.
* Properties built between 2011 and 2015 have the lowest average price.
There is no consistent relationship between the year a property was built and its listing price.
While the newest properties are among the most expensive, the oldest properties (1990-1995) have the highest
average prices. This suggests that year built alone is not a strong predictor of listing price in this dataset.
"""
print(interpretation)
```

### 13.What percentage of properties are sold above listing price?

```python
query13 = """
SELECT ROUND(
    SUM(CASE WHEN s.saleprice > l.price THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
    2
) AS percentage_sold_above_listing
FROM listings l
JOIN sales s ON l.listingid = s.listingid;
"""
result13 = run_query(query13)
display(result13)
interpretation = """
Approximately 49% of properties in the dataset were sold above their listing price,
indicating that nearly half of the properties experienced competitive bidding or strong demand that drove
the final sale price higher than the initial listing price.
"""
print(interpretation)
```

### 22.Does experience correlate with deals closed?

```python
query22 = """
SELECT
  CASE
    WHEN experienceyears <= 5  THEN '0-5 yrs'
    WHEN experienceyears <= 10 THEN '6-10 yrs'
    WHEN experienceyears <= 15 THEN '11-15 yrs'
    WHEN experienceyears <= 20 THEN '16-20 yrs'
    ELSE '21+ yrs'
  END AS experience_band,
  COUNT(*) AS num_agents,
  ROUND(AVG(dealsclosed),0) AS avg_deals
FROM agents
GROUP BY experience_band
ORDER BY MIN(experienceyears);
"""
result22 = run_query(query22)
display(result22)
agents = run_query("SELECT experienceyears, dealsclosed FROM agents")
print("correlation:", round(agents['experienceyears'].corr(agents['dealsclosed']), 3))
interpretation = """
The grouped analysis suggests that agents with more experience generally close more deals. 
Agents with 21+ years of experience have the highest average number of deals closed (166), while agents with 0–5 years of experience average 109 deals. 
Although there is a slight decline in the 16–20 years group, the overall pattern indicates that experienced agents tend to achieve better sales performance.

The correlation between experience and deals closed is 0.125, indicating a weak positive relationship. 
This suggests that while experience has some positive influence on sales performance, it is not a strong predictor. 
Other factors such as market conditions, client network, and individual agent effectiveness are likely to have a greater impact on the number of deals closed.
"""
print(interpretation)
```

### 26.What percentage of buyers are investors vs end users?


```python
query26 = """ 
SELECT buyertype,
       COUNT(*) AS num_buyers,
       ROUND(100.0 * COUNT(*) / (SELECT COUNT(*) FROM buyers), 2) AS percentage
FROM buyers
GROUP BY buyertype;
"""
result26 = run_query(query26)
display(result26)
interpretation = """ 
The buyer distribution is almost perfectly balanced between End Users and Investors. End Users account for 50.10% of all buyers, while Investors represent 49.90%. 
This indicates that the real estate market in this dataset attracts both owner-occupiers and investment buyers in nearly equal proportions.
"""
print(interpretation)  
```

### 27.Which cities have the highest loan uptake rate?

```python
query27 = """
SELECT l.city, SUM(b.loantaken) AS no_of_loans_taken, ROUND(100.0 * SUM(b.loantaken) / COUNT(*), 2) AS loan_uptake_rate
FROM buyers b
JOIN listings l ON l.listingid = b.saleid
GROUP BY l.city
ORDER BY loan_uptake_rate DESC;
"""
result27 = run_query(query27)
display(result27)
interpretation = """
Los Angeles has the highest loan uptake rate, with approximately 50.99% of buyers financing their property
purchases through loans. Phoenix has the lowest loan uptake rate at 49.50%. However, the variation across
cities is very small, with all cities having loan uptake rates close to 50%.
"""
print(interpretation)
```

### 28.What is the average loan amount by buyer type?

### 30.Do loan-backed purchases take longer to close?

```python
query30 = """
SELECT
    CASE
        WHEN b.loantaken = 1 THEN 'Loan Taken'
        ELSE 'No Loan'
    END AS loan_status,
    ROUND(AVG(s.daysonmarket), 2) AS average_days_on_market
FROM buyers b
JOIN sales s ON b.saleid = s.listingid
GROUP BY loan_status
ORDER BY average_days_on_market DESC;
"""
result30 = run_query(query30)
display(result30)
interpretation = """
Properties purchased using loans take slightly longer to close than properties purchased without loans.
On average, loan-backed purchases remain on the market for 62.15 days, compared to 61.07 days for non-loan
purchases.
"""
print(interpretation)
```
