# Database design 

```python
import os
if os.path.exists("project.db"):
    os.remove("project.db")
```

```python
conn = sqlite3.connect("project.db")
cursor = conn.cursor()
conn.execute("PRAGMA foreign_keys = ON;")
```

```python
def run_query(query):
    return pd.read_sql(query, conn)
    
```

```python
#Query to create the agents table
agents_table_query = """
CREATE TABLE IF NOT EXISTS agents(
    agentid TEXT PRIMARY KEY,
    name TEXT NOT NULL,
    phone TEXT NOT NULL,
    email TEXT NOT NULL,
    commissionrate REAL NOT NULL,
    dealsclosed INTEGER NOT NULL,
    rating REAL NOT NULL,
    experienceyears INTEGER NOT NULL,
    avgclosingdays INTEGER NOT NULL
);
"""
cursor.execute(agents_table_query)
agents_df.to_sql("agents", conn, if_exists="append", index=False)
conn.commit()
```

```python
agents_select_query = "select * from agents limit 1;"
run_query(agents_select_query)
```

```python
#Query to create the lisiting table
listings_table_query = """
CREATE TABLE IF NOT EXISTS listings(
    listingid TEXT PRIMARY KEY,
    city TEXT NOT NULL,
    propertytype TEXT NOT NULL,
    price REAL NOT NULL,
    sqft REAL NOT NULL,
    datelisted DATE NOT NULL,
    agentid TEXT NOT NULL,
    latitude REAL NOT NULL,
    longitude REAL NOT NULL,

    FOREIGN KEY(agentid)
        REFERENCES agents(agentid)
);
"""
cursor.execute(listings_table_query)
listings_df.to_sql("listings", conn, if_exists="append", index=False)
conn.commit()
```

```python
listings_select_query = "select * from listings limit 1;"
run_query(listings_select_query)
```

```python
#Query to create the sales table
sales_table_query = """
CREATE TABLE IF NOT EXISTS sales(
    listingid TEXT PRIMARY KEY,
    saleprice REAL NOT NULL,
    datesold DATE NOT NULL,
    daysonmarket INTEGER NOT NULL,

    FOREIGN KEY(listingid)
        REFERENCES listings(listingid)
);
"""
cursor.execute(sales_table_query)
sales_df.to_sql("sales", conn, if_exists="append", index=False)
conn.commit()
```

```python
sales_select_query = "select * from sales limit 1;"
run_query(sales_select_query)
```

```python
#Query to create the buyers table
buyers_table_query = """
CREATE TABLE IF NOT EXISTS buyers(
    buyerid INTEGER PRIMARY KEY,
    saleid TEXT NOT NULL,

    buyertype TEXT NOT NULL,
    paymentmode TEXT NOT NULL,
    loantaken BOOLEAN NOT NULL,
    loanprovider TEXT,
    loanamount REAL NOT NULL,

    FOREIGN KEY(saleid)
        REFERENCES sales(listingid)
);
"""
cursor.execute(buyers_table_query)
buyers_df.to_sql("buyers", conn, if_exists="append", index=False)
conn.commit()
```

```python
buyers_select_query = "select * from buyers limit 1;"
run_query(buyers_select_query)
```

```python
property_attributes_table_query = """
CREATE TABLE IF NOT EXISTS property_attributes(
    attributeid INTEGER PRIMARY KEY,
    listingid TEXT UNIQUE,
    bedrooms INTEGER NOT NULL,
    bathrooms INTEGER NOT NULL,
    floornumber INTEGER NOT NULL,
    totalfloors INTEGER NOT NULL,
    yearbuilt INTEGER NOT NULL,
    isrented BOOLEAN NOT NULL,
    tenantcount INTEGER NOT NULL,
    furnishingstatus TEXT NOT NULL,
    metrodistancekm REAL NOT NULL,
    parkingavailable BOOLEAN NOT NULL,
    powerbackup BOOLEAN NOT NULL,

    FOREIGN KEY(listingid)
        REFERENCES listings(listingid)
);
"""
cursor.execute(property_attributes_table_query)
property_df.to_sql("property_attributes", conn, if_exists="append", index=False)
conn.commit()
```

```python
property_select_query = "select * from property_attributes limit 1;"
run_query(property_select_query)
```

```python
# Agents
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_agents_rating
ON agents(rating);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_agents_experience
ON agents(experienceyears);
""")

# Listings
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_listings_city
ON listings(city);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_listings_propertytype
ON listings(propertytype);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_listings_price
ON listings(price);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_listings_agentid
ON listings(agentid);
""")

# Sales
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_sales_datesold
ON sales(datesold);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_sales_saleprice
ON sales(saleprice);
""")

# Buyers
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_buyers_saleid
ON buyers(saleid);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_buyers_buyertype
ON buyers(buyertype);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_buyers_paymentmode
ON buyers(paymentmode);
""")

# Property Attributes
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_property_bedrooms
ON property_attributes(bedrooms);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_property_yearbuilt
ON property_attributes(yearbuilt);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_property_metrodistance
ON property_attributes(metrodistancekm);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_property_listingid
ON property_attributes(listingid);
""")

conn.commit()

print("All indexes created successfully.")
```

## Property & Pricing Analysis

### 1. What is the average listing price by city?

### 2. What is the average price per square foot by property type?

```python
query2 = """
SELECT propertytype,ROUND(AVG(price / sqft), 2) AS avg_price_per_sqft FROM listings
GROUP BY propertytype
ORDER BY avg_price_per_sqft DESC;
"""
result2 = run_query(query2)
display(result2)
```

### 3. How does furnishing status impact property prices?

```python
query3 = """
SELECT p.furnishingstatus,ROUND(AVG(l.price), 2) AS average_price FROM listings l
JOIN property_attributes p ON l.listingid = p.listingid
GROUP BY p.furnishingstatus
ORDER BY average_price DESC;"""
result3 = run_query(query3)
display(result3)
interpretation = """
Interpretation
* Furnished properties have the highest average price.
* Semi-Furnished properties are almost identical in price (difference of only about 2,346).
* Unfurnished properties have the lowest average price.

Observation
The difference is very small:
Furnished vs Semi-Furnished -> ~2.3K
Furnished vs Unfurnished -> ~29.6K
"""
print(interpretation)
```

### 4. Do properties closer to metro stations command higher prices?

```python
query4 = """
SELECT
    CASE
        WHEN p.metrodistancekm <= 1 THEN 'Within 1 km'
        WHEN p.metrodistancekm <= 3 THEN '1 - 3 km'
        WHEN p.metrodistancekm <= 5 THEN '3 - 5 km'
        ELSE 'More than 5 km'
    END AS metro_distance,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY metro_distance
ORDER BY average_price DESC;"""
result4 = run_query(query4)
display(result4)
interpretation = """
Interpretation:
* Properties more than 5 km from a metro station have the highest average price.
* Properties within 1 km have the lowest average price.
The difference between the 1–3 km and 3–5 km groups is quite small.

Conclusion
No. This dataset does not indicate that properties closer to metro stations command higher prices. 
In fact, properties located more than 5 km away have the highest average listing price."""
print(interpretation) 
```

### 5. Are rented properties priced differently from non-rented ones?

```python
query5 = """
SELECT
    CASE
        WHEN p.isrented = 1 THEN 'Rented'
        ELSE 'Non Rented'
    END AS rental_status,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY rental_status
ORDER BY average_price DESC;
"""
result5 = run_query(query5)
display(result5)
interpretation = """
Interpretation:
The average prices of rented and non-rented properties are almost identical. 
This suggests that rental status alone does not have a significant impact on property prices in this dataset.
"""
print(interpretation)
```

### 6. How do bedrooms and bathrooms affect pricing?

```python
query6 = """
SELECT p.bedrooms,p.bathrooms,COUNT(*) AS total_properties,ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY p.bedrooms,p.bathrooms
ORDER BY p.bedrooms,p.bathrooms;
"""
result6 = run_query(query6)
display(result6) 
interpretation = """
The number of bedrooms and bathrooms alone does not appear to be a strong predictor of property price in this dataset. 
Property prices likely depend on a combination of factors such as city, property type, square footage, furnishing status, and location rather than just room counts."""
print(interpretation)
```

### 7. Do properties with parking and power backup sell at higher prices?

```python
query7 = """
SELECT
    CASE
        WHEN p.parkingavailable = 1 THEN 'Parking'
        ELSE 'No Parking'
    END AS parking,
    CASE
        WHEN p.powerbackup = 1 THEN 'Power Backup'
        ELSE 'No Power Backup'
    END AS power_backup,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY p.parkingavailable,p.powerbackup
ORDER BY average_price DESC;"""
result7 = run_query(query7)
display(result7)
interpretation = """
Interpretation:
* Properties with both parking and power backup have the highest average price.
* Properties with no parking and no power backup have the lowest average price.
Conclusion:
Properties that offer both parking and power backup tend to have the highest average prices. 
Although the price differences are modest, these amenities appear to add value and are associated with slightly higher-priced properties."""
print(interpretation)
```

### 8. How does year built influence listing price?

```python
query8 = """
SELECT
    CASE
        WHEN p.yearbuilt between 1990 and 1995  THEN 'Between 1990 and 1995'
        WHEN p.yearbuilt between 1996 and 2000  THEN 'Between 1996 and 2000'
        WHEN p.yearbuilt between 2001 and 2005  THEN 'Between 2001 and 2005'
        WHEN p.yearbuilt between 2006 and 2010  THEN 'Between 2006 and 2010'
        WHEN p.yearbuilt between 2011 and 2015  THEN 'Between 2011 and 2015'
        WHEN p.yearbuilt between 2016 and 2020  THEN 'Between 2016 and 2020'
        WHEN p.yearbuilt between 2021 and 2025  THEN 'Between 2021 and 2025'
        ELSE 'Other'
    END AS built_between,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY built_between
ORDER BY average_price DESC
"""
result8 = run_query(query8)
display(result8)
interpretation = """
Observations
* The oldest properties (1990–1995) have the highest average listing price.
* The newest properties (2021–2025) rank second.
* Properties built between 2011 and 2015 have the lowest average price.
There is no consistent relationship between the year a property was built and its listing price. 
While the newest properties are among the most expensive, the oldest properties (1990–1995) have the highest average prices. 
This suggests that year built alone is not a strong predictor of listing price in this dataset."""
print(interpretation)
```

### 9. Which cities have the highest average property prices?

### 10.How are properties distributed across price buckets?

```python
query10 = """
SELECT
    CASE
        WHEN price < 1000000 THEN 'Below 1M'
        WHEN price < 2000000 THEN '1M - 2M'
        WHEN price < 3000000 THEN '2M - 3M'
        WHEN price < 4000000 THEN '3M - 4M'
        ELSE 'Above 4M'
    END AS price_bucket,
    COUNT(*) AS total_properties
FROM listings
GROUP BY price_bucket
order by min(price);
"""
result10 = run_query(query10)
display(result10)
interpretation = """
The largest proportion of properties falls in the 1M–2M price range, with 4,800 properties."""
print(interpretation)
```

## Sales & Market Performance

### 11.What is the average days on market by city?

```python
query11 = """
SELECT l.city, ROUND(AVG(s.daysonmarket),0) as average_days_on_market
FROM listings l
join sales s on s.listingid = l.listingid
GROUP BY l.city
order by average_days_on_market desc;"""
result11 = run_query(query11)
display(result11)
interpretation = """
Los Angeles has the highest average days on market (65 days), indicating that properties take the longest to sell among the five cities.
In contrast, Houston has the lowest average (59 days), suggesting a relatively faster-moving real estate market."""
print(interpretation)
```

### 12.Which property types sell the fastest?

```python
query12 = """
SELECT l.propertytype, ROUND(avg(s.daysonmarket),0) as average_days_on_market
FROM listings l
join sales s on s.listingid = l.listingid
GROUP BY l.propertytype
order by average_days_on_market asc;"""
result12 = run_query(query12)
display(result12)
interpretation = """
Houses have the shortest average time on the market (58 days), making them the fastest-selling property type in the dataset. 
Condos take the longest to sell (67 days), while apartments and townhouses have similar average selling times of approximately 61 days. 
This suggests that property type has a moderate influence on how quickly a property is sold."""
print(interpretation)
```

### 13.What percentage of properties are sold above listing price?

```python
query13 = """
SELECT ROUND(SUM(
                CASE
                WHEN s.saleprice > l.price THEN 1
                ELSE 0
            END) * 100.0 / COUNT(*),
        2
    ) AS percentage_sold_above_listing
FROM listings l
JOIN sales s
ON l.listingid = s.listingid;"""
result13 = run_query(query13)
display(result13)
interpretation = """
Approximately 49% of properties in the dataset were sold above their listing price, 
indicating that nearly half of the properties experienced competitive bidding or strong demand that drove the final sale price higher than the initial listing price."""
print(interpretation)
```

```python
query30 = """
select CASE 
		WHEN b.loantaken = 1 THEN 'Loan Taken'
        ELSE 'No Loan'
        END AS loan_status,
		ROUND(AVG(s.daysonmarket),2 ) as average_days_on_market
from buyers b 
join sales s on b.saleid = s.listingid
group by loan_status
order by average_days_on_market desc;
"""
result30 = run_query(query30)
display(result30)
interpretation = """ 
Properties purchased using loans take slightly longer to close than properties purchased without loans. 
On average, loan-backed purchases remain on the market for 62.15 days, compared to 61.07 days for non-loan purchases.
"""
print(interpretation)
```
