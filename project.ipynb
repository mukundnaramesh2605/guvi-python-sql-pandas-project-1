# Database design 

```python
conn = sqlite3.connect("project.db")
cursor = conn.cursor()
conn.execute("PRAGMA foreign_keys = ON;")
```

```python
#Query to create the agents table
agents_table_query = """
CREATE TABLE agents(
    agentid TEXT PRIMARY KEY,
    name TEXT,
    phone TEXT,
    email TEXT,
    commissionrate REAL,
    dealsclosed INTEGER,
    rating REAL,
    experienceyears INTEGER,
    avgclosingdays INTEGER
);
"""
cursor.execute(agents_table_query)
agents_df.to_sql("agents", conn, if_exists="append", index=False)
conn.commit()
```

```python
agents_select_query = "select * from agents limit 1;"
result = pd.read_sql(agents_select_query, conn)
print(result)
```

```python
#Query to create the lisiting table
listings_table_query = """
CREATE TABLE listings(
    listingid TEXT PRIMARY KEY,
    city TEXT,
    propertytype TEXT,
    price REAL,
    sqft REAL,
    datelisted DATE,
    agentid TEXT,
    latitude REAL,
    longitude REAL,

    FOREIGN KEY(agentid)
        REFERENCES agents(agentid)
);
"""
cursor.execute(listings_table_query)
listings_df.to_sql("listings", conn, if_exists="append", index=False)
conn.commit()
```

```python
listings_select_query = "select * from listings limit 1;"
result = pd.read_sql(listings_select_query, conn)
print(result)
```

```python
#Query to create the sales table
sales_table_query = """
CREATE TABLE sales(
    listingid TEXT PRIMARY KEY,
    saleprice REAL,
    datesold DATE,
    daysonmarket INTEGER,

    FOREIGN KEY(listingid)
        REFERENCES listings(listingid)
);
"""
cursor.execute(sales_table_query)
sales_df.to_sql("sales", conn, if_exists="append", index=False)
conn.commit()
```

```python
sales_select_query = "select * from sales limit 1;"
result = pd.read_sql(sales_select_query, conn)
print(result)
```

```python
#Query to create the buyers table
buyers_table_query = """
CREATE TABLE buyers(
    buyerid INTEGER PRIMARY KEY,
    saleid TEXT,

    buyertype TEXT,
    paymentmode TEXT,
    loantaken BOOLEAN,
    loanprovider TEXT,
    loanamount REAL,

    FOREIGN KEY(saleid)
        REFERENCES sales(listingid)
);
"""
cursor.execute(buyers_table_query)
buyers_df.to_sql("buyers", conn, if_exists="append", index=False)
conn.commit()
```

```python
buyers_select_query = "select * from buyers limit 1;"
result = pd.read_sql(buyers_select_query, conn)
print(result)
```

```python
property_attributes_table_query = """
CREATE TABLE property_attributes(
    attributeid INTEGER PRIMARY KEY,
    listingid TEXT,
    bedrooms INTEGER,
    bathrooms INTEGER,
    floornumber INTEGER,
    totalfloors INTEGER,
    yearbuilt INTEGER,
    isrented BOOLEAN,
    tenantcount INTEGER,
    furnishingstatus TEXT,
    metrodistancekm REAL,
    parkingavailable BOOLEAN,
    powerbackup BOOLEAN,

    FOREIGN KEY(listingid)
        REFERENCES listings(listingid)
);
"""
cursor.execute(property_attributes_table_query)
property_df.to_sql("property_attributes", conn, if_exists="append", index=False)
conn.commit()
```

```python
property_select_query = "select * from property_attributes limit 1;"
result = pd.read_sql(property_select_query, conn)
print(result)
```

# SQL analysis 


### 1. What is the average listing price by city?

```python
query1 = """
SELECT city,ROUND(AVG(price), 2) AS average_listing_price
FROM listings
GROUP BY city
ORDER BY average_listing_price DESC;
"""
result1 = pd.read_sql(query1, conn)
result1
```

### 2. What is the average price per square foot by property type?

```python
query2 = """
SELECT propertytype,ROUND(AVG(price / sqft), 2) AS avg_price_per_sqft FROM listings
GROUP BY propertytype
ORDER BY avg_price_per_sqft DESC;
"""
result2 = pd.read_sql(query2, conn)
result2
```

### 3. How does furnishing status impact property prices?

```python
query3 = """
SELECT p.furnishingstatus,ROUND(AVG(l.price), 2) AS average_price FROM listings l
JOIN property_attributes p ON l.listingid = p.listingid
GROUP BY p.furnishingstatus
ORDER BY average_price DESC;"""
result3 = pd.read_sql(query3, conn)
display(result3)
interpretation = """
Interpretation
* Furnished properties have the highest average price.
* Semi-Furnished properties are almost identical in price (difference of only about 2,346).
* Unfurnished properties have the lowest average price.

Observation
The difference is very small:
Furnished vs Semi-Furnished -> ~2.3K
Furnished vs Unfurnished -> ~29.6K
"""
print(interpretation)
```

### 4. Do properties closer to metro stations command higher prices?

```python
query4 = """
SELECT
    CASE
        WHEN p.metrodistancekm <= 1 THEN 'Within 1 km'
        WHEN p.metrodistancekm <= 3 THEN '1 - 3 km'
        WHEN p.metrodistancekm <= 5 THEN '3 - 5 km'
        ELSE 'More than 5 km'
    END AS metro_distance,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY metro_distance
ORDER BY average_price DESC;"""
result4 = pd.read_sql(query4, conn)
display(result4)
interpretation = """
Interpretation:
* Properties more than 5 km from a metro station have the highest average price.
* Properties within 1 km have the lowest average price.
The difference between the 1–3 km and 3–5 km groups is quite small.

Conclusion
No. This dataset does not indicate that properties closer to metro stations command higher prices. 
In fact, properties located more than 5 km away have the highest average listing price."""
print(interpretation) 
```

### 5. Are rented properties priced differently from non-rented ones?

```python
query5 = """
SELECT
    CASE
        WHEN p.isrented = 1 THEN 'Rented'
        ELSE 'Non Rented'
    END AS rental_status,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY rental_status
ORDER BY average_price DESC;
"""
result5 = pd.read_sql(query5, conn)
display(result5)
interpretation = """
Interpretation:
The average prices of rented and non-rented properties are almost identical. 
This suggests that rental status alone does not have a significant impact on property prices in this dataset.
"""
print(interpretation)
```

### 6. How do bedrooms and bathrooms affect pricing?

```python
query6 = """
SELECT p.bedrooms,p.bathrooms,COUNT(*) AS total_properties,ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY p.bedrooms,p.bathrooms
ORDER BY p.bedrooms,p.bathrooms;
"""
result6 = pd.read_sql(query6, conn)
display(result6) 
```

### 7. Do properties with parking and power backup sell at higher prices?

```python
query7 = """
SELECT
    CASE
        WHEN p.parkingavailable = 1 THEN 'Parking'
        ELSE 'No Parking'
    END AS parking,
    CASE
        WHEN p.powerbackup = 1 THEN 'Power Backup'
        ELSE 'No Power Backup'
    END AS power_backup,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY p.parkingavailable,p.powerbackup
ORDER BY average_price DESC;"""
result7 = pd.read_sql(query7, conn)
display(result7)
interpretation = """
Interpretation:
* Properties with both parking and power backup have the highest average price.
* Properties with no parking and no power backup have the lowest average price.
Conclusion:
Properties that offer both parking and power backup tend to have the highest average prices. 
Although the price differences are modest, these amenities appear to add value and are associated with slightly higher-priced properties."""
print(interpretation)
```

### 8. How does year built influence listing price?

```python
query8 = """
SELECT
    CASE
        WHEN p.yearbuilt between 1990 and 1995  THEN 'Between 1990 and 1995'
        WHEN p.yearbuilt between 1996 and 2000  THEN 'Between 1996 and 2000'
        WHEN p.yearbuilt between 2001 and 2005  THEN 'Between 2001 and 2005'
        WHEN p.yearbuilt between 2006 and 2010  THEN 'Between 2006 and 2010'
        WHEN p.yearbuilt between 2011 and 2015  THEN 'Between 2011 and 2015'
        WHEN p.yearbuilt between 2016 and 2020  THEN 'Between 2016 and 2020'
        WHEN p.yearbuilt between 2021 and 2025  THEN 'Between 2021 and 2025'
    END AS built_between,
    COUNT(*) AS total_properties,
    ROUND(AVG(l.price), 2) AS average_price
FROM listings l
JOIN property_attributes p
ON l.listingid = p.listingid
GROUP BY built_between
ORDER BY average_price DESC
"""
result8 = pd.read_sql(query8, conn)
display(result8)
interpretation = """
Observations
* The oldest properties (1990–1995) have the highest average listing price.
* The newest properties (2021–2025) rank second.
* Properties built between 2011 and 2015 have the lowest average price.
There is no consistent relationship between the year a property was built and its listing price. 
While the newest properties are among the most expensive, the oldest properties (1990–1995) have the highest average prices. 
This suggests that year built alone is not a strong predictor of listing price in this dataset."""
print(interpretation)
```
