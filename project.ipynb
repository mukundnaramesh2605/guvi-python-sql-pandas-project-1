# Database design 

```python
conn = sqlite3.connect("project.db")
cursor = conn.cursor()
conn.execute("PRAGMA foreign_keys = ON;")
```

```python
#Query to create the agents table
agents_table_query = """
CREATE TABLE agents(
    agentid TEXT PRIMARY KEY,
    name TEXT,
    phone TEXT,
    email TEXT,
    commissionrate REAL,
    dealsclosed INTEGER,
    rating REAL,
    experienceyears INTEGER,
    avgclosingdays INTEGER
);
"""
cursor.execute(agents_table_query)
agents_df.to_sql("agents", conn, if_exists="append", index=False)
conn.commit()
```

```python
agents_select_query = "select * from agents limit 1;"
result = pd.read_sql(agents_select_query, conn)
print(result)
```

```python
#Query to create the lisiting table
listings_table_query = """
CREATE TABLE listings(
    listingid TEXT PRIMARY KEY,
    city TEXT,
    propertytype TEXT,
    price REAL,
    sqft REAL,
    datelisted DATE,
    agentid TEXT,
    latitude REAL,
    longitude REAL,

    FOREIGN KEY(agentid)
        REFERENCES agents(agentid)
);
"""
cursor.execute(listings_table_query)
listings_df.to_sql("listings", conn, if_exists="append", index=False)
conn.commit()
```

```python
listings_select_query = "select * from listings limit 1;"
result = pd.read_sql(listings_select_query, conn)
print(result)
```

```python
#Query to create the sales table
sales_table_query = """
CREATE TABLE sales(
    listingid TEXT PRIMARY KEY,
    saleprice REAL,
    datesold DATE,
    daysonmarket INTEGER,

    FOREIGN KEY(listingid)
        REFERENCES listings(listingid)
);
"""
cursor.execute(sales_table_query)
sales_df.to_sql("sales", conn, if_exists="append", index=False)
conn.commit()
```

```python
sales_select_query = "select * from sales limit 1;"
result = pd.read_sql(sales_select_query, conn)
print(result)
```

```python
#Query to create the buyers table
buyers_table_query = """
CREATE TABLE buyers(
    buyerid INTEGER PRIMARY KEY,
    saleid TEXT,

    buyertype TEXT,
    paymentmode TEXT,
    loantaken BOOLEAN,
    loanprovider TEXT,
    loanamount REAL,

    FOREIGN KEY(saleid)
        REFERENCES sales(listingid)
);
"""
cursor.execute(buyers_table_query)
buyers_df.to_sql("buyers", conn, if_exists="append", index=False)
conn.commit()
```

```python
buyers_select_query = "select * from buyers limit 1;"
result = pd.read_sql(buyers_select_query, conn)
print(result)
```

```python
property_attributes_table_query = """
CREATE TABLE property_attributes(
    attributeid INTEGER PRIMARY KEY,
    listingid TEXT,
    bedrooms INTEGER,
    bathrooms INTEGER,
    floornumber INTEGER,
    totalfloors INTEGER,
    yearbuilt INTEGER,
    isrented BOOLEAN,
    tenantcount INTEGER,
    furnishingstatus TEXT,
    metrodistancekm REAL,
    parkingavailable BOOLEAN,
    powerbackup BOOLEAN,

    FOREIGN KEY(listingid)
        REFERENCES listings(listingid)
);
"""
cursor.execute(property_attributes_table_query)
property_df.to_sql("property_attributes", conn, if_exists="append", index=False)
conn.commit()
```

```python
property_select_query = "select * from property_attributes limit 1;"
result = pd.read_sql(property_select_query, conn)
print(result)
```

```python
# Agents
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_agents_rating
ON agents(rating);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_agents_experience
ON agents(experienceyears);
""")

# Listings
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_listings_city
ON listings(city);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_listings_propertytype
ON listings(propertytype);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_listings_price
ON listings(price);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_listings_agentid
ON listings(agentid);
""")

# Sales
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_sales_datesold
ON sales(datesold);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_sales_saleprice
ON sales(saleprice);
""")

# Buyers
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_buyers_saleid
ON buyers(saleid);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_buyers_buyertype
ON buyers(buyertype);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_buyers_paymentmode
ON buyers(paymentmode);
""")

# Property Attributes
cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_property_bedrooms
ON property_attributes(bedrooms);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_property_yearbuilt
ON property_attributes(yearbuilt);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_property_metrodistance
ON property_attributes(metrodistancekm);
""")

cursor.execute("""
CREATE INDEX IF NOT EXISTS idx_property_listingid
ON property_attributes(listingid);
""")

conn.commit()

print("All indexes created successfully.")
```
